## Mengimport library

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import statistics as stats
from prettytable import PrettyTable
```

## Mengimport dataset

```python
df = pd.read_csv("studentsperformance.csv")
```

```python
df.head(10)
```

```python
df.shape
```

### Mengetahui jenis/type data variabel dalam dataset

```python
df.info()
```

```python
interval = pd.interval_range(0,100, freq=10)
frekuensi = pd.cut(df['reading_score'], interval)
frekuensi.value_counts(sort=False).rename_axis("reading_score").reset_index(name="frekuensi")
```

```python
df.hist(column='reading_score')
plt.xlabel("Nilai membaca")
plt.ylabel("Frekuensi")
plt.title("Reading Score")
```

```python
interval = pd.interval_range(0,100, freq=10)
frekuensi = pd.cut(df['writing_score'], interval)
frekuensi.value_counts(sort=False).rename_axis("writing_score").reset_index(name="frekuensi")
```

```python
df.hist(column='writing_score')
plt.xlabel("Nilai menulis")
plt.ylabel("Frekuensi")
plt.title("Writing Score")
```

### Menentukan nilai Mean, Min, Max

```python
df[['reading_score','writing_score']].describe()
```

### Menentukan median dan modus

```python
print("Median dari reading score dan writing score adalah ")
df[['reading_score','writing_score']].median()
```

```python
print("Modus dari reading_score dan writing score adalah ")
df[['reading_score','writing_score']].mode()
```

### Menentukan Varians

```python
print("Nilai varian reading score adalah ",df['reading_score'].var())
```

```python
print("Nilai varian writing score adalah ",df['writing_score'].var())
```

### Menentukan Jangkauan (Range)

Range = nilai max - nilai min

```python
df['reading_score'].max() - df['reading_score'].min()
```

```python
df['writing_score'].max() - df['writing_score'].min()
```

```python
tablerange = PrettyTable(['No','Variabel','Jangkauan(Range)'])
tablerange.add_row(['1','Reading score','83'])
tablerange.add_row(['2','Writing score','90'])
print(tablerange)
```

### Menentukan Kisaran Interkuartil

Q1,Q2,Q3 dari Reading Score

```python
print('Q1 : ',df['reading_score'].quantile(0.25))
print('Q2 : ',df['reading_score'].quantile(0.50))
print('Q3 : ',df['reading_score'].quantile(0.75))
```

Q1,Q2,Q3 dari Writing Score

```python
print('Q1 : ',df['writing_score'].quantile(0.25))
print('Q2 : ',df['writing_score'].quantile(0.50))
print('Q3 : ',df['writing_score'].quantile(0.75))
```

```python
df['reading_score'].quantile(0.75) - df['reading_score'].quantile(0.25)
```

```python
df['writing_score'].quantile(0.75) - df['writing_score'].quantile(0.25)
```

```python
table_kuartil = PrettyTable(['No','Variabel','Kisaran Interkuartil'])
table_kuartil.add_row(['1','Reading score','20.0'])
table_kuartil.add_row(['2','Writing score','21.25'])
print(table_kuartil)
```
